For the following potential energy for an anharmonic oscillator, we will create a figure that shows the angular frequency, $\omega$ versus the initial amplitude of the oscillator $Q_0$.

### The potential energy (notice the minus sign on the fourth-order term)

$$ \Delta U = \frac{1}{2} K Q^2 - \frac{1}{4} D Q^4 + \frac{1}{6} \Phi Q^6 $$

### Constrained parameters

Take $K = 1$, $D=2.4$, $\Phi=1$, and $M = 1$.

### Plotting constraints

Plot a range of initial displacements spanning $Q_0 = \pm 1.7$ (see the representative plot below)

### Extracting the angular frequency

Recall that in the second midterm and subsequent weeks we developed a strategy for a variant of this problem which used the following steps:

1. Solve the differential equation for $Q(t)$ for initial value $Q_0$ (take the initial velocity to be zero);
2. Use `ipywidgets` to explore this solution in a figure and extract approximate values of the period $T$ as a function of $Q_0$. We found $T$ by getting the time from peak to peak or valley to valley;
3. Interpolate the approximate values of $T$;
4. Use in interpolated values of $T$ as a guess for a minima finding algorithm;
5. Convert the extracted values of $T$ to angular frequency by $\omega=\frac{2\pi}{T}$.
6. To deal with the stationary solutions (e.g. $Q_0 = 0$), we used analytic results for the angular frequency by taking using the formula $\omega = \sqrt{\frac{1}{M}\frac{d^2\Delta U}{d Q^2}}$.

### Plot of the potential energy

```python
import numpy as np
import matplotlib.pyplot as plt

# define potential function
def potential_energy(Q, K=1.0, D=2.4, Phi=1.0):
    return (1/2) * K * Q**2 - (1/4) * D * Q**4 + (1/6) * Phi * Q**6

# define mesh for plotting
Q_mesh = np.linspace(-2,2,100)
U_mesh = potential_energy(Q_mesh)

# construct figure
plt.figure(figsize=(5, 4))
plt.plot(Q_mesh, U_mesh, color="gray")
plt.xlabel("$Q$")
plt.ylabel("$\Delta U$")
plt.xlim(-2,2)
plt.ylim(-0.1,0.25)
plt.show()
```

```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.optimize import minimize
from scipy.interpolate import interp1d
import ipywidgets as widgets

# Initialize Variables
K = 1
D = 2.4
Phi = 1
M = 1
V0 = 0

# Equation of Motion
def anharmonicOscillator(t, y, k, d, phi, m):
    Q, V = y
    dQdt = V
    dVdt = (-k*Q + d*Q**3 - phi*Q**5) / m
    return [dQdt, dVdt]

# Solve IVP
def solveSystem(q0, v0, k, d, phi, m):
    y0 = [q0, v0]
    sol = solve_ivp(anharmonicOscillator,
                    [0, 50], # You didn't specify a t range, but this was used in prior code examples
                    y0,
                    method='RK45',
                    t_eval=np.linspace(0, 50, 1000),
                    args=(k, d, phi, m))
    return sol # sol.t for t values, sol.y[0] and sol.y[1] for Q and V values respectively

#sol = solveSystem(0.01, V0, K, D, Phi, M) #Use me to view Q(t)!
#plt.plot(sol.t, sol.y[0])

# Evaluate Q(t)
def evaluateDisplacement(t, sol):
    #Every time I use weird numpy functions to find the index of where something is in a matching array, a fairy dies
    timeIndex = (np.abs(sol.t - t)).argmin() # https://stackoverflow.com/questions/2566412/find-nearest-value-in-numpy-array
    return sol.y[0][timeIndex]
    
# Extract and Graph Omega Values
def extractAngularFrequency(sol, g1, g2):
    firstMinima = minimize(evaluateDisplacement, g1, args=(sol)) 
    secondMinima = minimize(evaluateDisplacement, g2, args=(sol))
    period = np.sqrt((secondMinima.x - firstMinima.x)**2)
    return 2 * 3.1415926535 / period # np.pi() is a float so. lol.

# Interpolate Approximate Periods
approximatePeriods = [[-2, -1.5, -1, -0.5, -0.25, 0.01, 0.25, 0.5, 1, 1.5, 2], 
                      [3.5, 3.25, 3, 8, 7, 6, 7, 8, 3, 3.25, 3.5]] #0.01 used instead of 0 so I don't need to think about stationary case
QMesh = np.linspace(-1.7, 1.7, 1000)
linearInterpolation = interp1d(approximatePeriods[0], approximatePeriods[1], kind='linear')
interpolatedPeriods = linearInterpolation(QMesh)

# Create Angular Frequency Mesh (oof ow ouch my computation time)
omegaMesh = []
for i in QMesh:
    # This is a hideous solution, obvious by my initializing an empty ndarray. IDK how this was avoided in class, but solve_ivp refused to
    # take ndarrays, so I could not easily pass the QMesh to it to create a series of solutions. I could not find a single case of you doing
    # something similar in provided homework examples
    displacementValues = solveSystem(i, V0, K, D, Phi, M)
    periodIndex = np.where(QMesh == i)[0][0] # Each occurence should be the ONLY occurence.
    omega = extractAngularFrequency(displacementValues, interpolatedPeriods[periodIndex]*0.5, interpolatedPeriods[periodIndex]*1.5)
    omegaMesh.append(omega)

plt.figure(figsize=(10, 10))
plt.plot(QMesh.tolist(), omegaMesh, color="gray")
plt.xlabel("Initial Displacement (Arbitrary Units)")
plt.ylabel("Angular Frequency (Radians / Sec)")
plt.xlim(-2, 2)
plt.show()
```

Beyond these maximums, the plot slowly settles out towards a presumed asymptote. It looks something like a W, with the highest frequencies likely corresponding to relative maxima in potential energy.
