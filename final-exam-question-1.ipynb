```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import ipywidgets as widgets

#Initialize Variables
K1 = 1
K2 = 0.2
M1 = 1
M2 = 1
C = 0.5
Q10 = 0
Q20 = 1
V10 = 0
V20 = 0

# Plot Energy as Contour Plot
Q1Mesh = np.linspace(-3, 3, 1000)
Q2Mesh = np.linspace(-3, 3, 1000)
Q1Vals, Q2Vals = np.meshgrid(Q1Mesh, Q2Mesh)
U = 0.5 * K1 * Q1Vals**2 + 0.5 * K2 * Q2Vals**2 + (1/3) * C * Q1Vals * Q2Vals**2
plt.figure(figsize=(10, 10))
plt.contourf(Q1Vals, Q2Vals, U, levels=50, cmap='viridis')
plt.contour(Q1Vals, Q2Vals, U, levels=50, colors='white', linewidths=1)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.title("Potential Energy of Anharmonially Coupled Oscillators", fontsize=14)
plt.xlabel("Q1 Displacement (arbitrary units)", fontsize=12)
plt.ylabel("Q2 Displacement (arbitrary units)", fontsize=12)

# Define Equation of Motion
def anharmonicCoupledOscillators(t, y, m1, m2, k1, k2, c):
    Q1, Q2, V1, V2 = y
    dQ1dt = V1
    dQ2dt = V2
    dV1dt = (-k1 * Q1 - (1/3) * c * Q2**2) / m1
    dV2dt = (-k2 * Q2 - (2/3) * c * Q1 * Q2) / m2 
    return [dQ1dt, dQ2dt, dV1dt, dV2dt]

# Solve IVP
def findTrajectory(q10, q20, v10, v20, m1, m2, k1, k2, c):
    y0 = [q10, q20, v10, v20]
    sol = solve_ivp(anharmonicCoupledOscillators,
                    [0, 50], # Further beyond this begins to trace a very similar pattern. High values look really pretty!
                    y0,
                    method='RK45',
                    t_eval=np.linspace(0, 50, 200), 
                    args=(m1, m2, k1, k2, c))
    
    # Plot Q1, Q2
    plt.figure(figsize=(10, 10))
    plt.plot(sol.y[0], sol.y[1], color='red', label="Trajectory")
    plt.title("Trajectory of Anharmonically Coupled Oscillators", fontsize=14)
    plt.xlabel("Q1 Displacement (arbitrary units)", fontsize=12)
    plt.ylabel("Q2 Displacement (arbitrary units)", fontsize=12)

findTrajectory(Q10, Q20, V10, V20, M1, M2, K1, K2, C)
#findTrajectory(Q10, -1, V10, V20, M1, M2, K1, K2, C) #2nd case
#findTrajectory(1, 0, V10, V20, M1, M2, K1, K2, C) #3rd case
#findTrajectory(-1, 0, V10, V20, M1, M2, K1, K2, C) #4th case
plt.show()
```

The obvious difference between the 2 displacement variables is that, in the coupling term, Q1 is first order while Q2 is second order. Looking at the force applied to Q2, or, the partial derivative of U with respect to Q2, we can see that Q2 is in both terms:
<br>$F_{Q2} = -K_2Q_2 - \frac{2}{3}CQ_1Q_2$
<br>This means that, if Q2 starts with no displacement, no force is ever applied to Q2, and the oscillator behaves as though it was uncoupled and only a single anharmonic oscillator with respect to Q1. We can see that this is not true for Q1, with its force equation leaving it out of its coupling term:
<br>$F_{Q1} = -K_1Q_1 - \frac{1}{3}CQ_2^2$
In other words, displacement in Q2 can apply force on Q1, but displacement in Q1 **cannot** apply force on Q2. Using a physical example, we can think of applying force against the cylindrical face of a rod. While the squishing motion of the rod would displace linearly, the bending perpendicular to our squish would occur quadratically, creating an increasingly elliptical cross-section. In this case, Q1 would be our squishing motion, and Q2 our bending motion.
